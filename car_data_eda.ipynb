```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
```

```python
dataset = pd.read_csv('cars_delhi_eda.csv')
```

```python
dataset.head(1)
```

# variant

```python
print(dataset['variant'].value_counts()[:5])   # top 5
```

```python
index = dataset['variant'].value_counts()[:5].index
value = dataset['variant'].value_counts()[:5].values
plt.figure(figsize=(9,3))
plt.bar(index,value,color = "green")
plt.title("Top 5 Car variant")
plt.xlabel("variant")
plt.ylabel("counts of car")
plt.show()
```

# transmission

```python
dataset['transmission'].value_counts()
```

```python
transmission_index = dataset['transmission'].value_counts().index
transmission_value = dataset['transmission'].value_counts()[:5].values
plt.figure(figsize=(4,3))
plt.pie(transmission_value,labels=transmission_index,autopct="%0.2f%%",colors=["m","yellow"],shadow=True)
plt.title("Car transmission")
plt.legend(loc=1)
plt.show()
```

# owner_type

```python
dataset['owner_type'].value_counts()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x="owner_type",data=dataset,color="red")
plt.ylabel("count of car")
plt.title("no of car vs owner")
plt.show()
```

# fuel_type

```python
dataset['fuel_type'].value_counts()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x="fuel_type",data=dataset,color="red")
plt.ylabel("count of car")
plt.title("no of car vs fuel type")
plt.show()
```

# registeration

```python
dataset['registeration'].value_counts()[:5]
```

```python
index_registeration = dataset['registeration'].value_counts()[:5].index
value_registeration = dataset['registeration'].value_counts()[:5].values
plt.figure(figsize=(9,3))
plt.bar(index_registeration,value_registeration,color = "green")
plt.title("Top 5 Car registeration")
plt.xlabel("registeration")
plt.ylabel("counts of car")
plt.show()
```
